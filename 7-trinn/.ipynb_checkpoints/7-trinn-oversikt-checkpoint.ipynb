# Programmering i 7. trinn

## Kompetansemål
I kompetansemålet om programmering i 7. trinn står det:

Eleven skal kunne
* bruke programmering til å utforske data i tabeller og datasett

Her kan vi se en tydelig kobling til kompetansemålene i matematikk:
* logge, sortere, presentere og lese data i tabeller og diagrammer og begrunne valget av framstilling
* utforske og bruke hensiktsmessige sentralmål i egne og andres statistiske undersøkelser

## Forslag til læring
For å jobbe med kompetansemålet om utforsking av datasett anbefaler vi på det sterkeste å bruke tekstprogrammering. Det er mulig å jobbe noe med lister og tabeller i Scratch, men veldig begrenset hva man kan få til, og å lage beregninger på lister i Scratch er faktisk mer komplisert enn å gjøre det i tekst.

Her er det en god mulighet til å forsterke algoritmer, reglene og regnemåter fra statistikkemnet ved å oversette disse til programmering. Når eleven kan regne sentralmål på oversiktelige datasett, kan et dataprogram automatisere denne prosessen og gjøre samme beregninger på mye større datasett. Dette er selvfølgelig også mulig å gjøre i regneark, men da bruker man kun ferdiglagde kommandoer til å gjøre disse beregningene og man får ikke eierskap til algoritmen på samme måte som man gjør når de programmeres.

I tillegg er det mulig å bruke programmering til å visualisere data i forskjellige diagrammer, f.eks. søylediagram og kakediagram, men her må læreren vurdere om det er et hensiktsmessig verktøy for dette - eller om regneark er bedre egnet.

## Klasseromsaktiviteter
* Lage og sortere lister med tekst og tall
* Iterere (gå igjennom) en liste med for-sløyfe og finne sum og gjennomsnitt
* Sortere liste med tall og finne median i den sorterte listen
* Bruke innebygd listefunksjon for å finne typetall
