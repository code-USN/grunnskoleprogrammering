# Programmering i 5. trinn

## Kompetansemål
I kompetansemålene for 5. trinn kommer det tydelig fram at det skal programmeres og at det skal gjøres med datamaskin eller nettbrett. Det er i utgangspunktet ikke lenger nok med «bare» analog programmering.
Læringsmålet for 5. trinn er: Elevene skal kunne programmere algoritmer med bruk av variabler, vilkår og løkker. 
Forslag til læring
Her skal det programmeres, og nå skal datamaskinen eller nettbrettet fram og det skal brukes til koding. Vi introduserte og snakket om PRIMM som didaktisk metode og det er ingen tvil om at vi kan fortsette med dette. Blokkprogrammering vil være naturlig som inngang til programmering. Det finnes en rekke forskjellige apper og programmer til dette. Trinket, Scratch og MakeCode er nevnt i forrige trinn og vi tenker at det er helt naturlig å fortsette med samme verktøy som man brukte i sist gang (4. trinn). Poenget med dette er at det skal være gjenkjennbart og enkelt for både elver og lærer. 

Læringsmålet sier at elevene skal kunne programmere algoritmer med bruk av variabler, vilkår og løkker. De skal altså sette navn på variabler og gi disse variablene verdier i programmer de lager eller endrer (husk PRIMM). De skal også kunne programmere/ endre betingelser for disse og avgjøre om dette skal gjentas flere ganger (vilkår og løkker).  

Dette kan gjøres på flere måter, men det vil være meningsløst å gjøre dette fullstendig avskåret fra matematikken. Derfor er det viktig å koble programmeringsspråk, matematikkspråk og det daglige språket sammen. Dette kan gjøres ved at klassen lager et program som er nyttig. Man kan begynne å se til andre læringsmål innenfor matematikk for å se om det er noe man kan programmere som kan løse andre vanskelige ting i matematikken som man har på samme trinn. For eksempel dukker det opp et mål om å løse likninger gjennom logiske resonnementer. Det finnes også læringsmål knyttet til sannsynlighet som kan brukes. Vi har også mulighet for å se på en enkel omgjøringskalkulator for brøk og desimaltall, eller en kalkulator som forkort og utvider brøker.

Men husk at elevene her skal videreføre ideen som er plantet i tidligere trinn om at programmet skal være til hjelp og at det skal ha en funksjon i arbeidet deres. Vi skal ikke programmere bare for å programmere. Da vil elevene miste interesse for dette. Det må oppfattes som egennyttig.
