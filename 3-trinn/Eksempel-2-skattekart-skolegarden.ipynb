# Eksempel 2: Skattekart i skolegården

Lag en illustrasjon av skolegården og legg et rutenett over. I dette rutenettet bygger du nå opp et koordinatsystem. Her kan du både bruke negative og positive tall som i ordinære koordinatsystemer. Her kan du sette ut forskjellige premier til klassen som de må jobber for å finne. Her skal du som lærer oppgi koordinatene, mens elevene skal finne skattene. 

Denne aktiviteten kan også snus der elevene lager oppgaver for hverandre.
