# Eksempel 3: Geo-matte

Her skal vi bruke atlas eller virkelig kart på nettet (Atlas er bedre på grunn av rutenettet og inndelingen i breddegrader og lengdegrader).
Lag historier om personer som er ute på spennende oppdrag i verden, men at de har gått seg vill og ber klassen om hjelp for å komme hjem. Enten kan de oppgi egne posisjoner i koordinater og elevene skal finne ut hvor de er, eller de vi skal finne forteller hvor de er og elevene skal finne koordinatene til disse. Uansett skal de komme seg hjem til byen eller bygda der din skole er! Du kan øke vanskelighetsgraden ved at du sier at det er flere mennesker som skal hjem og at de skal plukkes opp i en spesiell rekkefølge!

Som i oppgavene på 1. og 2. trinn her kan du som lærer snu oppgaven slik at elevene lager oppgaver til hverandre.
