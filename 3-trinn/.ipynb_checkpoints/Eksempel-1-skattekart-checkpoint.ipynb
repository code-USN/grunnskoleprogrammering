# Eksempel 1: Skattekart i klasserommet

Klasserommet fullt av pulter er allerede et koordinatsystem, uten tall og origo. La veggen bak i klasserommet være x-aksen og veggen til venstre når du ser framover i klasserommet fra veggen bak i klasserommet være y-aksen. Bestem selv avstander og verdier på aksene – men det kan være lurt å bruke naturlige avstander som pultene til elevene allerede har skapt. Lag skattekart over rommet med koordinater og plasser ut skatter som elevene skal finne. Hva disse skattene skal være kan du bestemme selv, men det kan være nye kart som gi en ny oppgave! Her er poenget aktivitet og nøyaktighet.
