# Programmering i 3. trinn

## Kompetansemål

Kompetansemålet etter 3. trinn er en forlengelse eller utvidelse av kompetansemålet for 2. trinn. Kompetansemålet lyder:

Elevene skal kunne:
* lage og følge regler og trinnvise instruksjoner i lek og spill knyttet til koordinatsystemet

Dette målet må ses i sammenheng med et annet kompetansemål for trinnet: Elevene skal kunne eksperimentere med og forklare plasseringer i koordinatsystemet.

# Forslag til læring
Nå skal vi sette nesten den samme læringen inn i en mer matematisk kontekst. Det bør fortsatt ikke handle om at vi skal programmere med datamaskin eller nettbrett. Det kan fortsatt være snakk om UCP – men vi må begynne å rette fokus inn mot at det vi driver med, er matematikk.

Dersom vi setter kompetansemålet for programmering sammen med det andre målet med koordinatsystemet, kan vi bygge på det vi jobbet med innenfor programmering i 1. og 2. trinn. Her var det snakk om å gi instruksjoner og la elevene følge disse knyttet til hverdagen og overføre dette igjen til spill og lek. Nå kan vi anvende dette til det fulle ved å gå på skattejakt. Klasserommet med pultene kan bli et rutenett. Gymsalen med utsatte apparater kan bli et rutenett og skolegården kan bli et rutenett. Ved å bygge opp eller «eksperimentere med og forklare plasseringer i koordinatsystemet» kan elevene i programmering, eller i produksjon av algoritmer bli enda mer nøyaktige med tanke på at alle rom og plasser kan gis «adresser». Alle plasser i et koordinatsystem har tallkombinasjoner og dette kan utnyttes bed at man kan bygge ut og endre de algoritmene man har laget i første og andre trinn, eller man kan lage nye!

## Klasseromsaktiviteter
Vi husker fra 1. og 2. trinn at vi lagde algoritmer for blant annet å vaske hender, brannøvelse og innmarsj. Man kan nå se på om det er mulig å gjøre om de gamle algoritmene og forkorte disse fordi at vi har kunnskap om at steder i klasserommet og ute kan ha koordinater. Derfor kan det å bygge videre på allerede noe man har, være en gunstig start på det å forstå at koordinatsystem og «adresser» kan ha stor relevans for programmering og oppbygging av algoritmer. Men det finnes selvsagt nye oppgaver som kan gjøres. Under har vi laget noen forslag til klasseromsaktiviteter og uteaktiviteter som bygger videre på ideen om UCP – altså programmering uten pc eller iPad.
Alle aktivitetene vi har foreslått her krever en del forarbeid, men når maler og hoveddelene av oppgavene er laget, er det lett å ta de fram år etter år, gjenbruk vil skape mindre tid til forberedelser for de som kommer neste år.
