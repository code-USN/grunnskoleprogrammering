# Programmering i 6. trinn

## Kompetansemål

Her handler det om å kunne koble kunnskap opp mot geometriske figurer og mønstre. 


Kompetansemålet for trinnet lyder: 
* Elevene skal kunne bruke variabler, løkker, vilkår og funksjoner i programmering til å utforske geometriske figurer og mønstre.  

## Forslag til læring

Tidligere forslag har innebært bruk av PRIMM. Vi kan her ta steget videre og introdusere UMC - Use Modify Create og Par-programmering. 

### UMC 
Use Modify Create er forgjengeren til PRIMM. UMC legger mer vekt på et lengre prosjekt eller undervisningsopplegg. PRIMM er mer rettet mot undervisning i programmering i skolen, mens UMC er mer aktuelt for større programmer i feks simuleringer der programmet skal brukes og fungere. Læringen vil da skje over tid. 

### Par-programmering 
handler om å beholde dynamikken som finnes når man samarbeider og samtaler. Man går bort fra menneske-maskin-dialog og over til menneske-menneske-dialog. Parprogrammering er en samarbeidsmetode der to personer jobber sammen på én datamaskin. I programmeringsdidaktikk blir dette ofte sett på som et av de kraftigste verktøyene for å bygge forståelse, spesielt i matematikkfaget der logisk resonnering står sentralt. 

**Rollene: Pilot og Navigatør**

For at parprogrammering skal fungere didaktisk, må elevene ha tydelige roller som de bytter på ofte (for eksempel hvert 10. eller 15. minutt).
* Piloten: Dette er personen som sitter ved tastaturet. Sjåføren har fokus på de tekniske detaljene – å skrive koden, passe på syntaks, og utføre de umiddelbare instruksjonene.
* Navigatøren: Dette er personen som sitter ved siden av. Navigatøren har det "store bildet". Hen ser etter logiske feil, tenker på neste steg i algoritmen, og sjekker om koden faktisk løser det matematiske problemet. Hvorfor er dette bedre enn at alle sitter på hver sin PC?

### Analog programmering - UCP 
Ved å kombinere analog programmering og blokk/tekst programmering så gir det mulighet for at elevenes forståelse bedres. I analoge situasjoner kan utforsking av vinkler bedre elevenes forståelse av indre og ytre vinkler. 

Det er mange eksempler med både analog, blokk og tekstprogrammering som kan benyttes. Vi skal se på regulære mangelkanter, sirkler og forskjellige variasjoner av trekanter. Vi skal bruke PRIMM, PAR-programmering og **Parsonsproblems** i de følgende oppgavene. Vi skal også her introdusere turtle-geometri i tekstprogrammering, som i stor grad har de samme muligheter som scratch som får katten til å gå. Vi skal fokusere på at det skjer en transformasjon fra fokus på programmeringen til det matematikkfaglige. Slik vil da matematikken etterhvert komme mer og mer i fokus og dermed vil elevenes utforsking av matematikken være fokus.
