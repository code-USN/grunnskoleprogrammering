# Programmering i 4. trinn

## Kompetansemål

Etter fjerde trinn er det to kompetansemål som kan knyttes opp mot programmering.<br>Nå formaliseres kompetansemålene og de ligner mer på innholdet for en del andre kompetansemål innenfor matematikkfaget.  

* Det første kompetansemålet er: Elevene skal kunne utforske og beskrive strukturer og mønstre i lek og spill. 

* Det andre målet er: Elevene skal kunne lage algoritmer og uttrykke dem ved bruk av variabler, vilkår og løkker 

## Forslag til læring 

Spørsmålet her er om vi skal finne fram datamaskinen eller nettbrettet eller skal vi fortsette med UCP. Når det gjelder det første målet er det ingen tvil om at man ikke trenger PC. For å nå dette kompetansemålet kan man jobbe uten PC eller nettbrett. Når det kommer til det andre kompetansemålet som er å lage algoritmer og uttrykke dem, kan dette godt skje med penn og papir, men for at det skal kunne testes og sjekkes må de kjøres digitalt. Derfor er vår oppfordring at man her tar i bruk et digitalt verktøy.  

Her passer det å introdusere en didaktisk tilnærming som kalles PRIMM. Det er en forkortelse og står for predict – run – investigate – modify – make. Oversatt til norsk er det forutsi – kjør – undersøk – endre – produser. Det handler altså om at elever kan få en kode eller et program der de først forutsier hva koden vil gjøre. Deretter kjører de koden og ser om de hadde rett. Trinn tre blir jo da å undersøke hvorfor de fikk rett eller feil før de endrer koden slik at programmet gjør noe annet som elevene vil. Til slutt kan de bruke det de har lært gjennom de fire første stegene til å lage eller produsere sin egen kode. Her er det snakk om enkle programmer som gjør enkle ting og da helst i blokkprogrammer (Trinket, Scratch, MakeCode eller andre).
