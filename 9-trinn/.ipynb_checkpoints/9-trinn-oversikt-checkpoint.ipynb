# Programmering i 9. trinn

## Kompetansemål
Programmering i 9, trinn har en tydelig kobling til konkrete kompetansemål innen sannsynlighetsregning, å beregne sannsynlighet i spill og statistikk.
Eleven skal kunne:
* simulere utfall i tilfeldige forsøk og beregne sannsynligheten for at noe skal inntreffe, ved å bruke programmering

## Forslag til læring
Programmering kan her være et veldig bra verktøy til å supplere mer tradisjonelle aktivitere knyttet til sannsynlighet, som terningkast eller myntkast eller trekking av baller. Siden datamaskiner er i stand til å utføre beregninger veldig raskt, kan man utforske hva som skjer med et stokastisk forsøk når antall repetisjoner blir veldig stort (Store talls lov)

## Klasseromsaktiviteter
* Simulering av terningkast eller myntkast og se på relativ sannsynlighet i sammenheng med teoretisk sannsynlighet
* Simulering av "balltrekking" (uordnede utvalg uten tilbakelegging)
