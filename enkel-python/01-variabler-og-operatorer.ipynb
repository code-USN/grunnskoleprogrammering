# Variabler og operatorer (og utskrift)

## Variabler og variabeltyper

```python
heltall = 4
desimaltall = 3.14
tekst = "Hallo verden!"
boolsk = True

print(heltall)
print(type(heltall))
print(desimaltall)
print(type(desimaltall))
print(tekst)
print(type(tekst))
print(boolsk)
print(type(boolsk))
```

## Sette sammen tekst og tall med print()

```python
alder = 24
navn = "Thomas"
print( navn + " er " + str(alder) + " år gammel.")
print(navn, "er", alder, "år gammel")
```

## Operatorer (Regning)
Logiske operatorer på neste side

```python
a = 3
b = 5
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a**2)
```

## Sammensatt eksempel

```python
a = 3
b = 4
areal = a*b
print("Arealet av et rektangel med sidene " + str(a) + " og " + str(b) + " er " + str(areal))
```
