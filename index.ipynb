# Programmering for grunnskolen

På denne nettsiden vil vi etterhvert legge ut aktiviteter og opplegg som handler om programmering i skolen.

Disse nettsidene er ikke ment som kurs i grunnleggende programmering, men heller som en samling av kortere eller lengre eksempler på hvordan programmering kan brukes til undervisning og utforsking i matematikk. Dette er ment som hjelp til selvhjelp for lærere som underviser i matematikk i grunnskolen og som trenger litt ekstra «push» og hjelp for å bli bedre i programmering, eller å komme i gang på en god måte i klasserommet.

Vi har delt dette heftet inn etter klassetrinn som inneholder kompetansemål, tips til læring, og forslag til aktiviteter i klasserommet og oppgaver som du som lærer eller lærerstudent kan øve med, eller som du kan gi til klassen du har i det daglige. Hensikten med klasseromsaktivitetene er at det skal være enkle oppgaver som elever/ studenter eller lærere kan gjøre enkelt og samtidig få et lite innblikk i å bygge enkel programmeringskompetanse. 
Det viktigste å si innledningsvis er at programmering ikke skal være noe du har isolert i klasserommet noen uker, men at det er noe du faktisk bruker jevnlig som et verktøy for å systematiser eller forenkle hverdagen din!

De forskjellige aktivitetene har forskjellig didaktisk tilnærming, men ett viktig poeng er å lære seg å kunne lese kode og kunne koble koding til konkrete aktiviteter i matematikkfaget, og hvordan koding kan forsterke faglig forståelse og hvordan koding kan brukes til utforsking.

# Didaktisk tilnærming
Vi legger mer vekt på koblingen mellom sekvensforståelse, språk og logikk enn det å kunne bruke all mulige tekniske muligheter i programmering. Gjennom arbeid med eksemplene vil man lære seg teknikker og metoder for å kunne bruke programmering til å forsterke eller utforske fag. I eksemplene som er lagt opp til arbeid i par kan vi trekke inn PRIMM-metoden (Sue Sentance), men det er også mulig å jobbe med eksemplene alene (Use-Modify-Create).

Når man jobber med aktivitetene kan man selv velge om man vil kopiere og lime inn koden direkte, eller å skrive den av. Det er læring i å skrive av, men da er det større sjanse for å få kjørefeil hvis man skriver av feil. Dette kan være god trening i å feilsøke kode dersom elevene er mer vant til å jobbe med tekstprogrammering.
