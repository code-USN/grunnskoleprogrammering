# Programmering i 10. trinn

## Kompetansemål
I 10. trinn er det nå to aktuelle kompetansemål om programmering.

Eleven skal kunne:
* utforske matematiske egenskaper og sammenhenger ved å bruke programmering
* lese og forklare tekstbasert programmering (på høring)

Her er det ikke presisert noe matematisk innhold i kompetansemålet om programmering, og det gir en viss frihet for å utforske flere ting. Tidligere eksamensoppgaver har ofte handlet om geometri eller rekker.

## Forslag til læring
Siden kompetansemålet er veldig åpent kan det også være vanskelig å tolke hvis man er usikker på hvordan og når man kan bruke prorgammering i matematikkaktiviteter. En streng tolkning av kompetansemålet er at elevene nå skal være i stand til å bruke programmering på egen hånd, men dette er ofte ikke tilfellet. Det nye utkastet til læreplan i matematikk har også lagt inn en "bestilling" om at elevene skal være i stand til å lese og tolke tekstprogrammering. Vi mener dog  at dette er i seneste laget å begynne med tekst.

Programmering i 10. trinn kan kobles til mange andre matematiske temaer, som for eksempel
    * Polynomer
    * Stigningstall og vekstfart
    * Prosentvis endring (geometriske rekker)
    * Modellering
    * Analyse av datasett

## Klasseromsaktiviter
* Andregradskalkulator (symmetrilinje og ekstremalpunkt)
* Prosentvis endring, vekstfaktor (eksponentialfunksjoner, innskudd, lån. ol.)
    * Her kan man se på sum av rekker uten at man har den matematiske metoden
* Modellering
* Analyse
