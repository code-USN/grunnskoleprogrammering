## Forslag til læring
* Kompetansemålet er veldig åpent og vanskelig å tolke uten å vite hvordan man kan bruke programmering.
* Kompetansemålet kan tolkes som at elevene nå skal være i stand til å programmere en del på egen hånd
* Den nye utkastet legger nå inn en bestilling om at man skal bruke tekstprogrammering i det minste på ungdomsskolenivå
* Programmeringen kan kobles til mange andre kompetanser, som for eksempel
    * Polynomer
    * Stigningstall og vekstfart
    * Prosentvis endring (geometriske rekker)
    * Modellering
    * Analyse av datasett
