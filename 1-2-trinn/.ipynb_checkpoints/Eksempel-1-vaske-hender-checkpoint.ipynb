# Eksempel 1: Vaske hender

Elevene kan lage en plakat eller en liste der de påpeker korrekt måte å vaske hendene på. Her er det viktig at alle steg er med. (Å gå til vasken, stille seg ved vasken, skru på vannet osv) Når listen er komplett kan klassen vaske hendene etter oppskriften og man kan diskutere om alle trinnene er nødvendig eller om det er noe som mangler. Når man er enig om en felles oppskrift eller algoritme kan denne henges opp ved vasken!
