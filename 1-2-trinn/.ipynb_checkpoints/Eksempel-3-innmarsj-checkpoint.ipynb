# Eksempel 3: Innmarsj til klasserommet

Innmarsj til klasserommet En hverdagslig ting er å marsjere inn etter at skoleklokka ringer. Mange stiller opp to og to og går inn i klasserommet via garderoben der yttertøyet henges opp og sko settes på plass før de går videre. Alt dette kan også omformes til lister og plansjer.
