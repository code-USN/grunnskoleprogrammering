# Eksempel 2: Brannøvelse

Elevene kan lage sin egen plan for hvordan de skal «evakuere» når skolens brannalarm går. Her er det igjen viktig å få med seg alle gjøremål og alle rutiner som skal gjennomføres for oppstilling og telling av elever. Man kan bare gjennomføre deler eller gjøre hele. Igjen er det smart å ende opp med et produkt som en liste eller plansje for hvordan man skal gjennomføre den oppgaven eller det oppdraget man har. Heng gjerne produktet på veggen!
