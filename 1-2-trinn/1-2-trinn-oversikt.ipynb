# Programmering i 1. og 2. trinn

## Kompetansemål

Etter 2. trinn er det et kompetansemål som kan kobles til programmering og matematikk. Dette kompetansemålet er:

* Elevene skal kunne lage og følge regler og trinnvise instruksjoner i lek og spill

## Forslag til læring
Her snakker vi først og fremst om algoritmisk tenkning basert på det å gjennomføre kommandoer eller å gi nøyaktige og korrekte kommandoer. Algoritme er en liste over trinnvise prosedyrer for å fullføre en gitt oppgave. Dette kan være helt hverdagslige ting som barn blir bedt om å lage en algoritme for. Algoritmisk tenkning vil da kunne forklares som å strukturere det som skal gjøres i en god og fornuftig rekkefølge, eller sekvens, for å kunne fullføre en oppgave. Vi voksne kjenner dette igjen ved å finne oppskrifter på mat vi aldri har laget før. Da vil vi gjerne ha en forklaring steg-for-steg. Da vil matretten vi lager bli bra. Når vi har gjort den samme retten mange ganger, lager vi retten uten oppskrift. Sentrale begreper i dette arbeidet vil være sekvensielle ord (altså ord som sørger for rekkefølge), retningsord (ord som angir en retning), og løkker (noe som skal gjentas).

Unplugged Coding Practice (UCP) er et begrep som sier at vi ikke trenger å bruke PC for å kode, eller programmere. Vanlige ordinære oppgaver kan omgjøres til koding gjennom oppskrifter og direktiver. Retning, avstand, hastighet og varighet kan alle angis i slike oppskrifter. Også konkretene oppgaver kan beskrives. Slike oppgaver kan være å knyte skolisser, pusse tenner vaske hender eller smøre på en brødskive. Dette kan elever gjøre i par eller grupper. Egne og andre oppgaver kan selvsagt lages.

Hensikten er å få oppgaver som har konkret algoritmisk oppbygging og struktur. Videre kan man også se på om noe i egen «koding» eller algoritme mangler for at man skal få gjennomført oppgaven (debugging), eller man kan se om det er ting som er overflødig eller ikke trenger å være med for at oppgaven skal kunne bli gjennomført. Dersom elever sliter med skriving, kan det tegnes eller lages skilt for å kunne «vise» eller «lage» algoritmen!

## Klasseromsaktiviteter
I dette delkapittelet kommer vi til å legge fram noen enkle klasseromsaktiviteter som nødvendigvis ikke bør lages som eller presenteres som programmering. Her handler det om å gjøre barn oppmerksom på og kjent med at det lage en algoritme eller en sekvensiell oppskrift kan være fornuftig for å forstå nødvendigheten av å gjøre ting i riktig rekkefølge med presise beskrivelser. Oppgavene vi har valgt her kan være del av hverdagen, eller viktig informasjon som man må øve på i skolen. Eksempler på dette finner dere i menyen.

Lærere kan selv lage egne oppgaver som er knyttet til daglig gjøremål som klassen har. Det å bli vant til lister/ «oppskrifter» eller algoritmer vil være med på å trene barnas evne til å følge disse. Å flytte de inn i en hverdag vil gjøre de bedre til også å kunne følge dem i instruksjoner som spill og lek slik læringsmålet påpeker.
